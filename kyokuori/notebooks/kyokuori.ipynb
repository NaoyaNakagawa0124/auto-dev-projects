# 曲織 (kyokuori) — チャート曲を、 クラフトのパターンに織り直す

**目的**: DIY クラフター が、 今週のチャート上位曲を BPM / key / energy / valence / danceability / mode の 6 つの音楽特徴量から、 5 種類のクラフトパターン (Cross stitch / Knit / Quilt / Mosaic / Beading) に変換して、 次の作品のインスピレーションにする。

**仕組み**: パレットは 16 色の HSL から決定論的に生成される (key → hue、 valence → saturation、 energy → lightness)。 同じ曲なら 何度開いても 同じパターン。 ノートブックは 完全にオフライン、 Spotify / Billboard の API は 叩かない (シミュレーション値)。

**使い方**: 次のセルから順番に実行、 5 曲分の plate を表示。 気に入った パターン の グリッド を そのまま 図案 として 自分の 作品に 写す。

```python
import matplotlib.pyplot as plt
from kyokuori.tracks import top_n, by_id
from kyokuori.renderer import plate, palette_strip

plt.rcParams['figure.facecolor'] = '#faf6ee'
plt.rcParams['axes.facecolor']  = '#faf6ee'
plt.rcParams['savefig.facecolor'] = '#faf6ee'
```

## 今週の チャート上位 5 曲、 まず パレット だけ

曲ごとに 16 色 の HSL パレット。 これだけ 並べても、 曲 の 気分 が なんとなく 見えてくる。

```python
for t in top_n(5):
    fig = palette_strip(t)
    plt.show()
    plt.close(fig)
```

## 1 位: Espresso — 5 種類 の 織り目

BPM 104、 key D♯、 major、 energy 0.78、 valence 0.86。 明るく 弾む 曲。 Cross stitch では 対角線 が 強い モチーフ、 quilt では 三角分割 が 鮮やか、 mosaic は 12 個 の アンカー で 不規則 タイル。

```python
fig = plate(by_id('espresso'))
plt.show()
plt.close(fig)
```

## 2 位: APT. — 弾む 黄金 と 桃色

BPM 149、 key E、 major、 energy 0.85、 danceability 0.74。 色 が 飛び跳ねる 感じ、 quilt の 三角 が 賑やか。

```python
fig = plate(by_id('apt'))
plt.show()
plt.close(fig)
```

## 3 位: Fortnight — minor 調 の 抑え めの 紺 紫

BPM 94、 key A、 minor、 energy 0.42、 valence 0.31。 暗め の 落ち着いた パレット。 knit が 静か な ストライプ に なる。

```python
fig = plate(by_id('fortnight'))
plt.show()
plt.close(fig)
```

## 4 位: Good Luck, Babe! — 朱色 と キャメル の クロスステッチ

key C、 major、 energy 0.73、 valence 0.66。

```python
fig = plate(by_id('good-luck-babe'))
plt.show()
plt.close(fig)
```

## 5 位: Birds of a Feather — 中間 トーン の 上品 な パレット

key D、 major、 energy 0.51、 valence 0.45。 落ち着いた クラフト 向き。

```python
fig = plate(by_id('birds-of-a-feather'))
plt.show()
plt.close(fig)
```

## 自分の 作品に どう 持ち込む か

- **Cross stitch** の 対角線 → そのまま 図案 用 紙 に 写して 16 cm 角 の 刺繍。 1 マス = 1 ステッチ。
- **Knit** の 横ストライプ → 16 段 で 1 巡 する スカーフ や バッグ の ストライプ。
- **Quilt** の 4×4 三角 → 1 ブロック = 10 cm 角 で 縫って、 16 ブロック で 40 cm 角 の キルト。
- **Mosaic** → ガラス や 陶器 タイル を 12 色 で 用意、 anchor の 距離 で 割って 並べる。
- **Beading** → 16 段 × 16 列 で ビーズ織り、 偶数行 は 1 列 オフセット。

曲 の 雰囲気 を 残したまま、 自分 の 作品 に 翻訳 する 1 つ の 道。
