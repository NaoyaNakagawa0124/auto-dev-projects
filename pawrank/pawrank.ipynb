# 🏆 PawRank — Competitive Dog Stats Leaderboard

Every dog's a winner — but some are more winning than others.

This notebook ranks 24 dogs across 8 competitive categories with beautiful visualizations.

```python
from pawrank.dogs import get_dogs, CATEGORIES, CATEGORY_EMOJI, Dog, add_dog
from pawrank.rankings import overall_ranking, category_winners, trophy_room, rank_by_category, dog_percentile, head_to_head
from pawrank.charts import radar_chart, leaderboard_bar, overall_podium, category_heatmap, head_to_head_chart

dogs = get_dogs()
print(f'🐾 {len(dogs)} dogs loaded. Let the competition begin!')
```

## 🏆 Overall Champion Podium
Who's the top dog across all categories?

```python
overall_podium(dogs, top_n=5).show()
```

```python
overall = overall_ranking(dogs)
overall[['rank', 'name', 'breed', 'overall'] + CATEGORIES].head(10)
```

## 🏅 Trophy Room
Winners in every category:

```python
tr = trophy_room(dogs)
for _, row in tr.iterrows():
    emoji = CATEGORY_EMOJI.get(row['category'], '')
    print(f"{emoji} {row['category'].title():15s} 🏆 {row['winner']:12s} ({row['breed']}) — {row['score']}/100")
```

## 📊 Category Leaderboards
Top 10 in each category:

```python
for cat in CATEGORIES:
    leaderboard_bar(dogs, cat, top_n=10).show()
```

## 🕸️ Radar Comparison
Compare any dogs head-to-head:

```python
# Compare top 3 overall
top3 = overall_ranking(dogs).head(3)['name'].tolist()
radar_chart(dogs, top3, title=f'Top 3: {" vs ".join(top3)}').show()
```

```python
# Compare two specific dogs
radar_chart(dogs, ['Mochi', 'Shadow'], title='Mochi vs Shadow').show()
```

## ⚔️ Head to Head
Detailed comparison between two dogs:

```python
h2h = head_to_head(dogs, 'Ziggy', 'Rex')
print(f"\n⚔️ {h2h['dog1']} vs {h2h['dog2']}")
print(f"Result: {h2h['overall_winner']} wins {h2h['wins1']}-{h2h['wins2']}\n")
for cat, info in h2h['categories'].items():
    emoji = CATEGORY_EMOJI.get(cat, '')
    marker = '⭐' if info['winner'] == h2h['dog1'] else ('🔵' if info['winner'] == h2h['dog2'] else '➖')
    print(f"  {emoji} {cat:15s} {info['score1']:3d} vs {info['score2']:3d}  {marker} {info['winner']}")

head_to_head_chart(h2h).show()
```

## 📊 Full Scoreboard Heatmap
See every dog's stats at a glance:

```python
category_heatmap(dogs).show()
```

## 🐾 Add Your Dog!
Enter your dog's stats and see where they rank:

```python
# Uncomment and edit to add your own dog:
my_dog = Dog(
    name='Buddy',
    breed='Mixed',
    age=4,
    weight_kg=15,
    stats={
        'activity': 80,
        'tricks': 60,
        'obedience': 70,
        'cuteness': 100,  # obviously
        'friendliness': 90,
        'bravery': 50,
        'intelligence': 75,
        'fluffiness': 65,
    },
    owner='You',
)

dogs_with_mine = get_dogs() + [my_dog]
ranking = overall_ranking(dogs_with_mine)
my_rank = ranking[ranking['name'] == 'Buddy']['rank'].values[0]
total = len(dogs_with_mine)
print(f'\n🏆 Buddy ranks #{my_rank} out of {total} dogs!')
print(f'Overall score: {ranking[ranking["name"] == "Buddy"]["overall"].values[0]:.1f}/100')

# Percentile breakdown
pcts = dog_percentile(dogs_with_mine, 'Buddy')
for cat, pct in pcts.items():
    emoji = CATEGORY_EMOJI.get(cat, '')
    print(f'  {emoji} {cat:15s} Top {100-pct:.0f}%')

radar_chart(dogs_with_mine, ['Buddy', ranking.iloc[0]['name']], title='Your Dog vs #1').show()
```
