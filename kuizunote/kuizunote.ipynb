# 🧠 クイズノート — 脳トレパズル

**シニアのための楽しい脳トレーニング！**

10問のパズルに挑戦して、あなたの「脳タイプ」を診断しましょう。

### 遊び方
1. 各セルを上から順番に実行してください（Shift + Enter）
2. 問題を読んで、答えの数字を入力してください
3. 全問終わると脳タイプが診断されます

```python
# セットアップ（最初にこのセルを実行してください）
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt
import matplotlib.patches as patches
import numpy as np
from puzzles import *

# 日本語フォント設定
plt.rcParams['font.family'] = ['Hiragino Sans', 'Yu Gothic', 'Noto Sans CJK JP', 'sans-serif']
plt.rcParams['figure.facecolor'] = '#1a1a2e'
plt.rcParams['axes.facecolor'] = '#16213e'
plt.rcParams['text.color'] = '#e0e0e0'
plt.rcParams['axes.labelcolor'] = '#e0e0e0'
plt.rcParams['xtick.color'] = '#e0e0e0'
plt.rcParams['ytick.color'] = '#e0e0e0'

correct_answers = 0
total_questions = puzzle_count()
print('🧠 クイズノート準備完了！')
print(f'全{total_questions}問のパズルに挑戦しましょう。')
```

## 第1問: パターン認識 🔢

## 第4問: 図形パターン 🎨
下のグラフを見て答えてください。

```python
# 色パターンを描画
data = get_chart_data(4)
fig, ax = plt.subplots(1, 1, figsize=(10, 2))
for i, color in enumerate(data['colors']):
    rect = patches.FancyBboxPatch((i * 1.2, 0), 1, 1, boxstyle='round,pad=0.1',
                                   facecolor=color, edgecolor='white', linewidth=1)
    ax.add_patch(rect)
    ax.text(i * 1.2 + 0.5, 0.5, str(i + 1), ha='center', va='center',
            fontsize=14, color='white', fontweight='bold')

# Question mark
rect = patches.FancyBboxPatch((len(data['colors']) * 1.2, 0), 1, 1,
                               boxstyle='round,pad=0.1', facecolor='gray',
                               edgecolor='yellow', linewidth=2, linestyle='--')
ax.add_patch(rect)
ax.text(len(data['colors']) * 1.2 + 0.5, 0.5, '?', ha='center', va='center',
        fontsize=20, color='yellow', fontweight='bold')

ax.set_xlim(-0.5, (len(data['colors']) + 1) * 1.2 + 0.5)
ax.set_ylim(-0.5, 1.5)
ax.set_aspect('equal')
ax.axis('off')
ax.set_title(data['title'], fontsize=14, pad=10)
plt.tight_layout()
plt.show()

p = get_puzzle(4)
print(f"\n{p['question']}\n")
ans = int(input('あなたの答え: '))
if check_answer(4, ans):
    correct_answers += 1
    print('✅ 正解！すばらしい！')
else:
    print(f'❌ 残念... 正解は {p["answer"]} でした')
print(f'💡 {p["explanation"]}')
```

## 第9問: 鏡パターン 🪞

```python
p = get_puzzle(9)
# Visualize the mountain pattern
seq = [1, 2, 3, 4, 5, 4, 3, 2, None]
fig, ax = plt.subplots(figsize=(8, 4))
x_vals = list(range(len(seq)))
y_plot = [v if v else 0 for v in seq]
colors = ['#6c5ce7' if v else '#ffd700' for v in seq]
bars = ax.bar(x_vals, y_plot, color=colors, edgecolor='white', linewidth=0.5)
for i, v in enumerate(seq):
    label = str(v) if v else '?'
    c = '#e0e0e0' if v else '#ffd700'
    ax.text(i, (v or 1) + 0.2, label, ha='center', fontsize=16, color=c, fontweight='bold')
ax.set_ylim(0, 7)
ax.set_title('数列のパターン', fontsize=14, pad=10)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
ax.spines['left'].set_color('#555')
ax.spines['bottom'].set_color('#555')
plt.tight_layout()
plt.show()

print(f"\n{p['question']}\n")
ans = int(input('あなたの答え: '))
if check_answer(9, ans):
    correct_answers += 1
    print('✅ 正解！すばらしい！')
else:
    print(f'❌ 残念... 正解は {p["answer"]} でした')
print(f'💡 {p["explanation"]}')
```

## 🏆 結果発表！

```python
score = calculate_score(correct_answers, total_questions)
brain = get_brain_type(score)

print('=' * 50)
print('🧠 クイズノート — 結果発表')
print('=' * 50)
print(f'\n正解数: {correct_answers} / {total_questions}')
print(f'スコア: {score}点\n')
print(f'あなたの脳タイプ: {brain["type"]}')
print(f'{brain["desc"]}\n')

# Score visualization
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 5))

# Gauge chart
theta = np.linspace(0, np.pi, 100)
ax1.fill_between(np.cos(theta), np.sin(theta), 0, color='#16213e')
score_angle = np.pi * (1 - score / 100)
ax1.plot([0, np.cos(score_angle) * 0.8], [0, np.sin(score_angle) * 0.8],
         color='#ff6b6b', linewidth=3)
ax1.set_xlim(-1.2, 1.2)
ax1.set_ylim(-0.2, 1.2)
ax1.set_aspect('equal')
ax1.axis('off')
ax1.text(0, -0.1, f'{score}点', ha='center', fontsize=24, fontweight='bold', color='#ffd700')
ax1.set_title('スコアメーター', fontsize=14, pad=10)

# Category breakdown
types = ['パターン', '数列', '論理', 'グラフ']
type_scores = [0, 0, 0, 0]
type_totals = [0, 0, 0, 0]
# Simplified: just show overall
ax2.barh(['正解', '不正解'], [correct_answers, total_questions - correct_answers],
         color=['#6c5ce7', '#555'], edgecolor='white', linewidth=0.5)
ax2.set_xlim(0, total_questions + 1)
ax2.set_title('正解 vs 不正解', fontsize=14, pad=10)
ax2.spines['top'].set_visible(False)
ax2.spines['right'].set_visible(False)
ax2.spines['left'].set_color('#555')
ax2.spines['bottom'].set_color('#555')

plt.tight_layout()
plt.show()

print('\nお疲れさまでした！また挑戦してくださいね 🎉')
```
