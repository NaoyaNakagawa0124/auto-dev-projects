# 季 重ね 帖 (kisanae-chou)

**1 年 分 の アニメ 視聴 履歴 を、 日本 の 季節 (旬 / 二十四節気 / 花) に 重ねる 美 ノート**

下 の セル を 上 から 順 に 実行 する と、 `out/` ディレクトリ に 4 種 の PNG (合計 15 枚) が 生成 さ れ ます。

- `year_timeline.png` — 12 月 横並び の 重ね 表
- `month_01.png` ... `month_12.png` — 月 別 拡大 カード
- `season_scroll.png` — 春夏秋冬 縦 巻物
- `palette_year.png` — 12 月 パレット

## セットアップ

````python
import matplotlib
matplotlib.use('Agg')
import matplotlib.pyplot as plt

from kisanae_chou.data import SAMPLE_ANIME, SEASONAL_EVENTS
from kisanae_chou.plots import (
    year_timeline_fig, month_card_fig, season_scroll_fig, palette_year_fig, save_fig
)

YEAR = 2026
print(f'アニメ: {len(SAMPLE_ANIME)} 本、 季節 event: {len(SEASONAL_EVENTS)} 件')
````

## 1. 年 表 — 全 12 月 を 横 に 並べる

アニメ の 放映 タイミング を 色 円 で、 旬 / 節気 / 花 を 帯 で 表示。

````python
fig = year_timeline_fig(SAMPLE_ANIME, SEASONAL_EVENTS, YEAR)
save_fig(fig, 'out/year_timeline.png')
print('out/year_timeline.png 保存 完了')
````

## 2. 月 別 カード — 12 枚

1 月 から 12 月 まで、 各 月 を 拡大 し て アニメ と 旬 を 並べた Insta 風 1200×1200 カード。

````python
for m in range(1, 13):
    fig = month_card_fig(SAMPLE_ANIME, SEASONAL_EVENTS, YEAR, m)
    save_fig(fig, f'out/month_{m:02d}.png')
print('12 枚 保存 完了')
````

## 3. 春夏秋冬 縦 巻物

1 季 = 1 段、 4 段 を 縦 に 並べた 巻物 風 1 枚。

````python
fig = season_scroll_fig(SAMPLE_ANIME, SEASONAL_EVENTS, YEAR)
save_fig(fig, 'out/season_scroll.png')
print('out/season_scroll.png 保存 完了')
````

## 4. 月 別 パレット 帯

12 月 を 1 色 ずつ 並べた 横長 帯。

````python
fig = palette_year_fig(YEAR)
save_fig(fig, 'out/palette_year.png')
print('out/palette_year.png 保存 完了')
````
